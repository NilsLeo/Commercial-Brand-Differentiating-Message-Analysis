```python
cont_music_major_dict = {
    'AD0254': '0',
    'AD0256': '1',
    'AD0259': '0',
    'AD0315': '0',
    'AD0318': '0',
    'AD0320': '0',
    'AD0374': '0',
    'AD0377': '0',
    'AD0380': '0',
    'AD0411': '0',
    'AD0416': '1',
    'AD0431': '0',
    'AD0484': '1',
    'AD0491': '0',
    'AD0503': '0',
    'AD0527': '0',
    'AD0530': '0',
    'AD0545': '0',
    'AD0600': '0',
    'AD0602': '0',
    'AD0608': '0',
    'AD0640': '0',
    'AD0644': '1',
    'AD0651': '0',
    'AD0691': '0',
    'AD0694': '1',
    'AD0704': '0',
    'AD0769': '0',
    'AD0772': '0',
    'AD0785': '0'
}
```

```python
for year in os.listdir(input_folder):
  years.append(year)
years.sort()

for year in years:
  ad_year = os.path.join(input_folder, year)
  for ad in os.listdir(ad_year):
    ad_path = os.path.join(ad_year, ad)
    # Used for an empty column to divide the ads
    idx += 1
    name_empty_column = ' ' * idx

    # Get the value from the dictionary
    ad_name = ad[0:6]
    cont_music = cont_music_dict.get(ad_name, '')
    cont_music_major = cont_music_major_dict.get(ad_name, '')

    df_2 = pd.read_excel(ad_path)
    df_1 = pd.concat([df_1, df_2], axis=1)
    df_1[f'cont_music_{ad}'] = cont_music
    df_1[f'cont_music_major_{ad}'] = cont_music_major
    df_1[f'Conclusion_Text_{ad}'] = ''
    df_1[f'Conclusion_Stimmung_Umwandlung_Video_{ad}'] = ''
    df_1[f'{name_empty_column}'] = ''
```

```python
df_1.to_excel('/content/drive/MyDrive/SuperBowl_Project_FUB/Überprüfung_Mood/all_ads_Intervall_25.xlsx', index=False)
```
